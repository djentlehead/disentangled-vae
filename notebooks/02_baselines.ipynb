```python
import os
import sys
print(f"The current working directory is: {os.getcwd()}")


sys.path.append('..')
print("Parent directory added to path. Python can now see the 'src' folder.")

from src.data.midi_dataset import MIDIDataset
from torch.utils.data import DataLoader


TOKENIZED_DATA_DIR = "../data/processed/tokenized"
BATCH_SIZE = 4

dataset = MIDIDataset(data_dir=TOKENIZED_DATA_DIR, max_sequence_length=512)
print(f"✅ Found {len(dataset)} tokenized files.")

dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)
print(f"✅ Created DataLoader with batch size {BATCH_SIZE}.")

try:
    x_batch, y_batch = next(iter(dataloader))

    print("\n--- Inspecting one batch ---")
    print(f"Shape of input batch (x): {x_batch.shape}")
    print(f"Shape of target batch (y): {y_batch.shape}")
    print(f"Data type of input: {x_batch.dtype}")
    print("This confirms the data is loaded and shaped correctly for training!")
    
except (StopIteration, FileNotFoundError) as e:
    print(f"Error: {e}. Is the TOKENIZED_DATA_DIR correct and non-empty?")
```
